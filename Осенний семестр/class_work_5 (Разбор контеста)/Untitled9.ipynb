```python
import pandas as pd
from datetime import date
df = pd.read_csv('/content/weatherHistory.csv.zip')
df["Formatted Date"]=pd.to_datetime(df['Formatted Date'], utc=True)
df["Formatted Date"].dt.month
print(df[df["Formatted Date"].dt.month.isin([6, 6, 8])].agg({'Wind Speed (km/h)': 'mean', "Temperature (C)": "mean"}),'-ЛЕТО')
print(df[df["Formatted Date"].dt.month.isin([1, 2, 12])].agg({'Wind Speed (km/h)': 'mean', "Temperature (C)": "mean"}),'-ЗИМА')
print(df[df["Formatted Date"].dt.month.isin([3, 4, 5])].agg({'Wind Speed (km/h)': 'mean', "Temperature (C)": "mean"}),'-ВЕСНА')
print(df[df["Formatted Date"].dt.month.isin([9, 10, 11])].agg({'Wind Speed (km/h)': 'mean', "Temperature (C)": "mean"}),'-ОСЕНЬ')

```
