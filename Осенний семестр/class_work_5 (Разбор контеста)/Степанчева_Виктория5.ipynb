```python
from datetime import date
df["Formatted Date"]=pd.to_datetime(df['Formatted Date'], utc=True)
df['Date'] = df["Formatted Date"].dt.date
a = df[df['Date']==date(2016,9,9)]
a['Temperature (C)'].mean()
```

```python
df["Formatted Date"] = pd.to_datetime(df["Formatted Date"], utc=True)
print('лето:')
print(df[df["Formatted Date"].dt.month.isin([6, 7, 8])].agg({'Wind Speed (km/h)': 'mean', 'Temperature (C)': "mean"}))
print('осень:')
print(df[df["Formatted Date"].dt.month.isin([9, 10, 11])].agg({'Wind Speed (km/h)': 'mean', 'Temperature (C)': "mean"}))
print('зима:')
print(df[df["Formatted Date"].dt.month.isin([1, 2, 12])].agg({'Wind Speed (km/h)': 'mean', 'Temperature (C)': "mean"}))
print('весна:')
print(df[df["Formatted Date"].dt.month.isin([3, 4, 5])].agg({'Wind Speed (km/h)': 'mean', 'Temperature (C)': "mean"}))
```
