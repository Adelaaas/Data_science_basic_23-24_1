````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('https://raw.githubusercontent.com/Adelaaas/Data_science_basic_23-24_1/main/%D0%9E%D1%81%D0%B5%D0%BD%D0%BD%D0%B8%D0%B9%20%D1%81%D0%B5%D0%BC%D0%B5%D1%81%D1%82%D1%80/work_%D0%B2%D0%B8%D0%B7%D1%83%D0%B0%D0%BB%D0%B8%D0%B7%D0%B0%D1%86%D0%B8%D1%8F/train_var_1.csv')
````

````python
"1"
plt.figure(figsize=(10, 5))
plt.scatter(df['Vmag'], df['e_Plx'])
plt.xlabel('Vmag')
plt.ylabel('e_Plx')
plt.title('Диаграмма разброса')
"Вывод: очень много значений от 1,5 до 10,5 по Vmag и от 0.5 до 1 по e_Plx"
````

````python
plt.figure(figsize=(10, 5))
cmap = sns.cubehelix_palette(as_cmap=True, light=.9)
sns.heatmap(df.isna().transpose(), cmap=cmap,
            cbar_kws={'label': 'Missing Data'})
"Вывод: цвет один, и он показывает 0, значит пропусков нет"
````

````python
"3"
fig, axes = plt.subplots(nrows=1, ncols=7)
sns.boxplot(df["Vmag"], ax=axes[0], width=1)
sns.boxplot(df["Plx"], ax=axes[1], width=1)
sns.boxplot(df["e_Plx"], ax=axes[2], width=1)
sns.boxplot(df["B-V"], ax=axes[3], width=1)
#sns.boxplot(df["SpType"], ax=axes[4], width=1)
sns.boxplot(df["Amag"], ax=axes[5], width=1)
sns.boxplot(df["TargetClass"], ax=axes[6], width=1)
axes[0].set_xlabel("Vmag")
axes[1].set_xlabel("Plx")
axes[2].set_xlabel("e_Plx")
axes[3].set_xlabel("B-V")
axes[4].set_xlabel("SpType")
axes[5].set_xlabel("Amag")
axes[6].set_xlabel("TargetClass")
plt.show()
"Вывод: Plx много выбросов снизу и сверху; e-plx много выбросов сверху; B-V много выбросов сверху; Amag много выбросов снизу и сверху; TargetClass нет выбросов"
````

````python
"4"
plt.figure(figsize=(10, 5))
sns.barplot(x=df['TargetClass'], y=df['TargetClass'].value_counts())
plt.title('barplot по TargetClass')
plt.xlabel('TargetClass')
plt.ylabel('Количество')
"Равное большое количство значений у 1 и 0"
````

````python
"5"
numeric_columns = df.select_dtypes(include=["number"])
plt.figure(figsize = (10, 5))
sns.heatmap(df.select_dtypes(include=["number"]).corr(), xticklabels=df.select_dtypes(include=["number"]).corr().columns, yticklabels=df.select_dtypes(include=["number"]).corr().columns, cmap='RdYlGn', center=0, annot=True)
plt.title("Матрица корреляции для всех переменных")
plt.show()
"Вывод: самая маленькая корреляция у B-V и TargetClass, самая большая у Plx и Amag"
````

````python
"6"
plt.figure(figsize=(10,5))
sns.pairplot(df)
plt.title("графики парных зависимостей между всеми числовыми переменными.")
plt.show()
"Только у TargetClass зависимость с переменными по бокам"
````

````python
"7"
plt.figure(figsize=(13, 8))

sns.histplot(df['TargetClass'], bins=30, kde=True, color='skyblue')

plt.title('Распределение переменной TargetClass')

plt.xlabel('TargetClass')
plt.ylabel('Частота')
"Вывод: основная частота между 0-0.2 и 0.8-1"
````
