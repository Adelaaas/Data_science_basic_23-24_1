```python
import pandas as pd
import matplotlib.pyplot as plt
link = 'https://raw.githubusercontent.com/Adelaaas/Data_science_basic_23-24_1/main/%D0%9E%D1%81%D0%B5%D0%BD%D0%BD%D0%B8%D0%B9%20%D1%81%D0%B5%D0%BC%D0%B5%D1%81%D1%82%D1%80/class_work_6%20(%D0%92%D0%B8%D0%B7%D1%83%D0%B0%D0%BB%D0%B8%D0%B7%D0%B0%D1%86%D0%B8%D1%8F)/train_var_2.csv'
df = pd.read_csv(link)
plt.scatter(df["age"], df["balance"])
plt.xlabel("age")
plt.ylabel("balance")
plt.title("Диаграмма разброса age и balance")
plt.show()
```

balance не зависит от age

```python
import seaborn as sns
sns.heatmap(df.isna().transpose(), cbar_kws={'label': 'Пропущенные значения'})
```

Пропущенных значений нет

```python

plt.figure(figsize=(30,10))
sns.boxplot(df["age"], ax=plt.subplot(1, 7, 1))
sns.boxplot(df["balance"], ax=plt.subplot(1, 7, 2))
sns.boxplot(df["day"], ax=plt.subplot(1, 7, 3))
sns.boxplot(df["duration"], ax=plt.subplot(1, 7, 4))
sns.boxplot(df["campaign"], ax=plt.subplot(1, 7, 5))
sns.boxplot(df["pdays"], ax=plt.subplot(1, 7, 6))
sns.boxplot(df["previous"], ax=plt.subplot(1, 7, 7))

```
