```python
import pandas as pd
import matplotlib.pyplot as plt
import random
import seaborn as sns
import numpy as np
x = [18, 13, 19, 14, 2, 11, 2, -30, 10, 1, 8, 11, 18, 15, 50]
y = [7, 9, 19, 13, 2, 17, 20, -40, 9, 10, 18, 3, 2, 2, 80]
z = [0.203, 0.443, 0.603, 0.879, 0.871, 0.164, 0.078, 0.766, 0.353, 0.949, 0.391, 0.2, 0.094, 0.692, 0.992]
plt.figure(figsize=(10, 10))
plt.subplot(2, 2, 1)
plt.grid()
plt.title("линейный график y(x)")
plt.xlabel("Это x")
plt.ylabel("Это y")
plt.plot(x, y)
plt.subplots_adjust(wspace=0.3, hspace=0.3)

plt.subplot(2, 2, 2)
plt.title("точечный график y(x)")
plt.xlabel("Это x")
plt.ylabel("Это y")
plt.scatter(x, y)

plt.subplot(2, 2, 3)
plt.title("гистограмма расределения x")
plt.xlabel("Значения x")
plt.ylabel("Частота")
plt.hist(x)

plt.subplot(2, 2, 4)
plt.title("гистограмма расределения y")
plt.xlabel("Значения y")
plt.ylabel("Частота")
plt.hist(y)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import random
import seaborn as sns
import numpy as np
x = [18, 13, 19, 14, 2, 11, 2, -30, 10, 1, 8, 11, 18, 15, 50]
y = [7, 9, 19, 13, 2, 17, 20, -40, 9, 10, 18, 3, 2, 2, 80]
z = [0.203, 0.443, 0.603, 0.879, 0.871, 0.164, 0.078, 0.766, 0.353, 0.949, 0.391, 0.2, 0.094, 0.692, 0.992]
b = pd.DataFrame()
b["x"] = x
b["y"] = y
b["z"] = z
sns.heatmap(b.corr(), xticklabels=b.corr().columns, yticklabels=b.corr().columns, cmap='RdYlGn', center=0, annot=True)
```

Наибольшая корелляция между x и y, наименьшая между x и z
