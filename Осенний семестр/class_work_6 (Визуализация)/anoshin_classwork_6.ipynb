```python
import pandas
import pandas as pd
import matplotlib.pyplot as plt
import random
import seaborn as sns
import numpy as np

x = [18, 13, 19, 14, 2, 11, 2, -30, 10, 1, 8, 11, 18, 15, 50]
y = [7, 9, 19, 13, 2, 17, 20, -40, 9, 10, 18, 3, 2, 2, 80]
z = [0.203, 0.443, 0.603, 0.879, 0.871, 0.164, 0.078, 0.766, 0.353, 0.949, 0.391, 0.2, 0.094, 0.692, 0.992]

plt.figure(figsize=(10, 7))

plt.subplot(2, 2, 1)
plt.title('Линейный график', fontsize=22)
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x, y)


plt.subplot(2, 2, 2)
plt.title('Точечный график', fontsize=22)
plt.xlabel("x")
plt.ylabel("y")
plt.scatter(x, y, c="r")

plt.subplot(2, 2, 3)
plt.title('Гистограма для х', fontsize=22)
plt.xlabel("x")
plt.ylabel("кол-во")
plt.hist(x)

plt.title('Гистограма для у', fontsize=22)
plt.subplot(2, 2, 4)
plt.xlabel("y")
plt.ylabel("кол-во")
plt.hist(y)

plt.show()

# Первые 4 графика должны быть на одной figure, но на разных subplots
# 1 - обычный линей график y(x)
# 2 - точечный график y(x)
# 3 - гистограмма расределения x
# 4 - гистограмма распределения y

plt.figure(figsize=(5, 5))
plt.title('Ящик с усами для x', fontsize=22)
plt.ylabel('Values')
plt.boxplot(x)
plt.show()

plt.figure(figsize=(5, 5))

plt.title('Ящик с усами для y', fontsize=22)
plt.ylabel('Values')
plt.boxplot(y)
plt.show()

plt.figure(figsize=(5, 5))

# 5 и 6 графики должны быть на разных фигурах
# 5 - boxplot для x
# 6 - boxplot для y

df = pandas.DataFrame({"x": x, "y": y, "z": z})

sns.heatmap(df.corr(), xticklabels=df.corr().columns, yticklabels=df.corr().columns, cmap='RdYlGn', center=0,
            annot=True)
plt.title('Correlogram of xyz', fontsize=22)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
# 7 график:
# создайте датафрейм со столбцами x, y, z
# постройте heatmap корреляции между x, y, z

# Все графики должны быть подписаны

print("На первом графике зависимости нет")
print(
    "На втором графике 2 значения находятся в разных концах графика. Большиство сосредоточено в середине между -20 и 20")
print("На третьем графике одно значение <-20, одно > 40. Больше всгео значений между 0 и 20 - шесть.")
print(
    "На четвером графике одно значение между -40 и -20, одно между 60 и 80. Больше всгео значений между 0 и 30 - 7.")
print("Пятый графике такой же как и третий")
print("Шестой графике такой же как и четвертый")
print("На седьмом графике x зависит от y. Остальное нет.")
```
