```python
import pandas as pd
import matplotlib.pyplot as plt
import random
import seaborn as sns
import numpy as np

x = [18, 13, 19, 14, 2, 11, 2, -30, 10, 1, 8, 11, 18, 15, 50]
y = [7, 9, 19, 13, 2, 17, 20, -40, 9, 10, 18, 3, 2, 2, 80]
z = [0.203, 0.443, 0.603, 0.879, 0.871, 0.164, 0.078, 0.766, 0.353, 0.949, 0.391, 0.2, 0.094, 0.692, 0.992]

# Первые 4 графика должны быть на одной figure, но на разных subplots
# 1 - обычный линей график y(x)
# 2 - точечный график y(x)
# 3 - гистограмма расределения x
# 4 - гистограмма распределения y

# 5 и 6 графики должны быть на разных фигурах
# 5 - boxplot для x
# 6 - boxplot для y

# 7 график:
# создайте датафрейм со столбцами x, y, z
# постройте heatmap корреляции между x, y, z

# Все графики должны быть подписаны
```

```python
fig, ax = plt.subplots(2,2)
ax[0,0].set_title("линейный график y(x)")
ax[0,0].set_xlabel("x")
ax[0,0].set_ylabel("y")
ax[0,1].set_title("точечный график y(x)")
ax[0,1].set_xlabel("x")
ax[0,1].set_ylabel("y")
ax[1,0].set_title("гистограмма расределения x")
ax[1,0].set_xlabel("x")
ax[1,0].set_ylabel("Кол-во вхождений")
ax[1,1].set_title("гистограмма распределения y")
ax[1,1].set_xlabel("y")
ax[1,1].set_ylabel("Кол-во вхождений")
ax[0,0].plot(x, y, c = "pink")
ax[0,1].scatter(x, y, c = 'mediumslateblue')
ax[1,0].hist(x, color = 'lightcoral', ec = 'firebrick')
ax[1,1].hist(y, color = 'lightblue', ec = 'blue')
plt.subplots_adjust(hspace = 0.6, wspace = 0.6)
plt.show()
```

```python
data = []
data.append(x)
data.append(y)
data.append(z)
df = pd.DataFrame(data).transpose ()
df.columns = ("x","y","z")
print(df)


```

```python
plt.figure(figsize=(8,6), dpi= 80)
sns.heatmap(df.corr(), xticklabels=df.corr().columns, yticklabels=df.corr().columns, cmap='RdPu', center=0, annot=True)
```
