```python
import pandas
import pandas as pd
import matplotlib.pyplot as plt
import random
import seaborn as sns
import numpy as np

x = [18, 13, 19, 14, 2, 11, 2, -30, 10, 1, 8, 11, 18, 15, 50]
y = [7, 9, 19, 13, 2, 17, 20, -40, 9, 10, 18, 3, 2, 2, 80]
z = [0.203, 0.443, 0.603, 0.879, 0.871, 0.164, 0.078, 0.766, 0.353, 0.949, 0.391, 0.2, 0.094, 0.692, 0.992]

plt.figure(figsize=(10, 7))

plt.subplot(2, 2, 1)
plt.title(' ', fontsize=22)
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x, y)


plt.subplot(2, 2, 2)
plt.title(' ', fontsize=22)
plt.xlabel("x")
plt.ylabel("y")
plt.scatter(x, y, c="r")

plt.subplot(2, 2, 3)
plt.title('  ', fontsize=22)
plt.xlabel("x")
plt.ylabel("-")
plt.hist(x)

plt.title('  ', fontsize=22)
plt.subplot(2, 2, 4)
plt.xlabel("y")
plt.ylabel("-")
plt.hist(y)

plt.show()

#  4      figure,    subplots
# 1 -    y(x)
# 2 -   y(x)
# 3 -   x
# 4 -   y

plt.figure(figsize=(5, 5))
plt.title('    x', fontsize=22)
plt.ylabel('Values')
plt.boxplot(x)
plt.show()

plt.figure(figsize=(5, 5))

plt.title('    y', fontsize=22)
plt.ylabel('Values')
plt.boxplot(y)
plt.show()

plt.figure(figsize=(5, 5))

# 5  6      
# 5 - boxplot  x
# 6 - boxplot  y

df = pandas.DataFrame({"x": x, "y": y, "z": z})

sns.heatmap(df.corr(), xticklabels=df.corr().columns, yticklabels=df.corr().columns, cmap='RdYlGn', center=0,
            annot=True)
plt.title('Correlogram of xyz', fontsize=22)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
# 7 :
#     x, y, z
#  heatmap   x, y, z

#     

print("    ")
print(
    "   2      .      -20  20")
print("     <-20,  > 40.     0  20 - .")
print(
    "      -40  -20,   60  80.     0  30 - 7.")
print("      ")
print("      ")
print("   x   y.  .")
```
