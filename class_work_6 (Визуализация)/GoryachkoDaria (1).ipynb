```python
import pandas as pd
import matplotlib.pyplot as plt
import random
import seaborn as sns
import numpy as np

x = [18, 13, 19, 14, 2, 11, 2, -30, 10, 1, 8, 11, 18, 15, 50]
y = [7, 9, 19, 13, 2, 17, 20, -40, 9, 10, 18, 3, 2, 2, 80]
z = [0.203, 0.443, 0.603, 0.879, 0.871, 0.164, 0.078, 0.766, 0.353, 0.949, 0.391, 0.2, 0.094, 0.692, 0.992]
```

```python
plt.figure(figsize=(12, 7))


plt.subplot(2, 2, 1)
plt.title('обычный линей график y(x)', fontsize=10)
plt.plot(x, y)


plt.subplot(2, 2, 2)
plt.title('точечный график y(x)', fontsize=10)
plt.scatter(x, y)


plt.subplot(2,2,3)
plt.title('гистограмма расределения x', fontsize=10)
plt.hist(x)


plt.subplot(2,2,4)
plt.title('гистограмма расределения y', fontsize=10)
plt.hist(y)

```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x)
plt.title('boxplot для x', fontsize=10)
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(y)
plt.title('boxplot для y', fontsize=10)
```

```python
#датафрейм со столбцами x,y,x
data = []
data.append(x)
data.append(y)
data.append(z)
df = pd.DataFrame(data).transpose ()
df.columns = ("X","Y","Z")
print(df)
```

```python
plt.figure(figsize=(10, 7))
sns.heatmap(df.corr(), center=0, annot = True)
plt.title('heatmap корреляции между x, y, z', fontsize=10)
```
