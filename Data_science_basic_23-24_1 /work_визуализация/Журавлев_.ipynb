```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/Adelaaas/Data_science_basic_23-24_1/main/work_%D0%B2%D0%B8%D0%B7%D1%83%D0%B0%D0%BB%D0%B8%D0%B7%D0%B0%D1%86%D0%B8%D1%8F/train_var_2.csv')
df.head()

fig, ax = plt.subplots()
x = df["age"]
y = df["balance"]
plt.title("График зависимости balance от age")
plt.xlabel("age") # подись для оси х
plt.ylabel("balance") # подись для оси y
ax.scatter(x, y, c = 'deeppink')
```

```python
import numpy as np
import seaborn as sns

x = np.shape(df)[1]
y = df.isna().sum()


plt.figure(figsize=(15,7))

cmap = sns.cubehelix_palette(as_cmap=True, light=.9)

sns.heatmap(df.isna().transpose(), cmap=cmap,
            cbar_kws={'label': 'Missing Data'})

plt.title("График анализа значений NaN")
plt.ylabel("столбцы") # подись для оси х
plt.xlabel("cтроки") # подись для оси y
fig.set_facecolor('floralwhite')

plt.show()
```

```python
numerical_variables = df.select_dtypes(include=['int64', 'float64']).columns
plt.figure(figsize=(15, 10))

for i, var in enumerate(numerical_variables, 1):
    plt.subplot(3, 3, i)
    plt.boxplot(df[var])
    plt.title(f'{var}')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
df['deposit'].value_counts().plot(kind='bar', color=['skyblue'])
plt.title('Barplot для переменной deposit')
plt.xlabel('deposit')
plt.ylabel('Количество')
plt.show()
```

5

```python
plt.figure(figsize=(12, 8))
x = df.corr()
plt.imshow(x, cmap='coolwarm', aspect='auto')
plt.colorbar()
plt.title('Матрица корреляции для всех переменных')
plt.xticks(range(len(x)), x.columns, rotation=90)
plt.yticks(range(len(x)), x.columns)
plt.show()
```
