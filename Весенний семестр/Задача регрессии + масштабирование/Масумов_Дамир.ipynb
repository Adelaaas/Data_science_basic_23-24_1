```python
def cn(value):
  if value == 'gas':
    return 0
  elif value == 'diesel':
    return 1
  else:
    return value
def asp(value):
  if value == 'std':
    return 0
  elif value == 'turbo':
    return 1
  else:
    return value
def cb(value):
  if value == 'sedan':
    return 0
  elif value == 'hatchback':
    return 1
  elif value == 'convertible':
    return 2
  elif value == 'wagon':
    return 3
  elif value == 'hardtop':
    return 4
  else:
    return value
def dw(value):
  if value == 'rwd':
    return 0
  elif value == 'fwd':
    return 1
  elif value == '4wd':
    return 2
  else:
    return value
def el(value):
  if value == 'front':
    return 0
  elif value == 'rear':
    return 1
  else:
    return value
def et(value):
  if value == 'ohc':
    return 0
  elif value == 'ohcv':
    return 1
  elif value == 'ohcf':
    return 2
  elif value == 'dohc':
    return 3
  elif value == 'l':
    return 4
  elif value == 'rotor':
    return 5
  elif value == 'dohcv':
    return 6
  else:
    return value
def cln(value):
  if value == 'four':
    return 0
  elif value == 'five':
    return 1
  elif value == 'six':
    return 2
  elif value == 'eight':
    return 3
  elif value == 'three':
    return 4
  elif value == 'two':
    return 5
  elif value == 'twelve':
    return 6
  else:
    return value
def fsy(value):
  if value == '2bbl':
    return 0
  elif value == 'mpfi':
    return 1
  elif value == 'mfi':
    return 2
  elif value == 'spfi':
    return 3
  elif value == 'idi':
    return 4
  elif value == '4bbl':
    return 5
  elif value == '1bbl':
    return 6
  elif value == 'spdi':
    return 7
  else:
    return value

df['fueltype'] = df['fueltype'].apply(cn)
df['aspiration'] = df['aspiration'].apply(asp)
df['carbody'] = df['carbody'].apply(cb)
df['drivewheel'] = df['drivewheel'].apply(dw)
df['enginelocation'] = df['enginelocation'].apply(el)
df['enginetype'] = df['enginetype'].apply(et)
df['cylindernumber'] = df['cylindernumber'].apply(cln)
df['fuelsystem'] = df['fuelsystem'].apply(fsy)
df
```

```python
X = df.drop(columns=['price'])
Y = df['price']
df.head()
```

обозначаем результат как y, а призника как x

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)

print(f'Train dataset size: {X_train.shape}, {y_train.shape}')
print(f'Test dataset size: {X_test.shape}, {y_test.shape}')
```

80% отводим на тренироку программы, а 20% на тесты

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
model = LinearRegression()
model.fit(X_train, y_train)
```

Создадим модель линейной регрессии, Обучим линейную регрессию на тренировочном стандартизированном наборе данных и а этапе обучения алгоритм рассчитывает веса признаков и строит линейное уравнение регрессии

```python
print('Веса всех признаков (w1, ..., w13): ', model.coef_)
print('Свободный коэффицент уравнения w0:', model.intercept_)
```

получили вес и коэффицент уравнения

Создадим массив названия признаков

```python
features = X_train.columns

coeff_df = pd.DataFrame(model.coef_, columns=['Coefficient'])
coeff_df['features'] = features

coeff_df.sort_values(by='Coefficient')
```

сделаем предсказание

```python
y_pred = model.predict(X_test)

print('Mean Absolute Error:', mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', mean_squared_error(y_test, y_pred))
print('R2 score:', r2_score(y_test, y_pred))
```
