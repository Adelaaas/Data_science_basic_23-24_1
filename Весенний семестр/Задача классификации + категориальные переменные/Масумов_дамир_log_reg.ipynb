```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
# Ð’Ñ‹Ð´ÐµÐ»Ð¸Ð¼ Ð²ÐµÐºÑ‚Ð¾Ñ€ Ð¿Ñ€Ð¸Ð·Ð½Ð°ÐºÐ¾Ð² Ð¥ Ð¸ Ð²ÐµÐºÑ‚Ð¾Ñ€ Ð¾Ñ‚Ð²ÐµÑ‚Ð¾Ð² y
X = df.drop(['target'], axis=1)
y = df['target']

# train - Ð½Ð° Ñ‚Ñ€ÐµÐ½Ð¸Ñ€Ð¾Ð²Ð¾Ñ‡Ð½Ñ‹Ðµ Ð¸ Ñ‚ÐµÑÑ‚Ð¾Ð²Ñ‹Ðµ
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LogisticRegression()
# model = KMeans()
# model = LinearRegression()

# Ð¼Ð¾Ð´ÐµÐ»ÑŒ ÑƒÑ‡Ð¸Ñ‚ÑÑ, Ð´Ð»Ñ ÑÑ‚Ð¾Ð³Ð¾ ÐµÐ¹ Ð½ÑƒÐ¶Ð½Ñ‹ Ð¿Ñ€Ð°Ð²Ð¸Ð»ÑŒÐ½Ñ‹Ðµ
model.fit(X_train, y_train)

# Ñ‚ÐµÑÑ‚Ð¸Ñ€ÑƒÐµÐ¼, Ð´Ð»Ñ Ð¿Ñ€Ð¸Ð·Ð½Ð°ÐºÐ¾Ð² Ñƒ ÐºÐ¾Ñ‚Ð¾Ñ€Ñ‹Ñ… Ð½ÐµÑ‚ Ð¾Ñ‚Ð²ÐµÑ‚Ð¾Ð², Ð¼Ñ‹ Ð¿Ñ€ÐµÐ´ÑÐºÐ°Ð·Ñ‹Ð²Ð°ÐµÑ‚ Ð¾Ñ‚Ð²ÐµÑ‚
answers_pred = model.predict(X_test)
```
