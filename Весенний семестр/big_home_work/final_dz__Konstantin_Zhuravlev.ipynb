В данной ситуации будет логично заполнить 0 все NaN, будем считать отсутствие информации о механиках, количестве пользователей и доменов за их отсутствие. ID вообще нет смысла учитывать при линейной регрессии так что пусть будет там вместо NaN будет -1.

```python
df['Domains'] = df['Domains'].fillna('0')
df['Owned Users'] = df['Owned Users'].fillna(0)
df['Mechanics'] = df['Mechanics'].fillna('0')
df['Year Published'] = df['Year Published'].fillna(0)
df['ID'] = df['ID'].fillna('-1')

s = df.isna().sum()
print(s)
```

```python
df.dtypes
```

Надо перевести столбец с нашей целевой переменной в численный тип данных, для этого с помощбю волшебной функции мы сначала приводим всё в нужный формат и тогда уже переводим.

```python
def clean_numeric(column):
    column = column.str.strip()
    column = column.str.replace(',', '.')
    column = column.str.replace('[^0-9.]', '', regex=True)
    return column

df['Complexity Average'] = clean_numeric(df['Complexity Average'])
df['Rating Average'] = clean_numeric(df['Rating Average'])


df['Complexity Average'] = pd.to_numeric(df['Complexity Average'], errors = 'coerce')
df['Rating Average'] = pd.to_numeric(df['Rating Average'], errors = 'coerce')

df0 = df

df
```

Теперь кодируем категориальные переменные.

```python
df_gt = df['Mechanics'].str.get_dummies(sep=', ')
df = pd.concat([df.drop(columns=['Mechanics']), df_gt], axis=1)

df_gt = df['Domains'].str.get_dummies(sep=', ')
df = pd.concat([df.drop(columns=['Domains']), df_gt], axis=1)

df
```

Сейчас делим датасет на две части чтобы можно было делать регрессию.

```python
not_nes = ['ID', 'Name', 'Rating Average']

x = df.loc[:, ~df.columns.isin(not_nes)]
y = df['Rating Average']

x = x.loc[:, ~x.columns.duplicated()]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
x.shape
```

Теперь обучаем и тестируем нашу модель внутри тренировочной части, смотрим на приблизительные результаты.

```python
model = LinearRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

# print('Веса всех признаков (x1, ..., x16): ', model.coef_)
# print('Свободный коэффицент: ', model.intercept_)

print('Mean Absolute Error: ', mean_absolute_error(y_test, y_pred))
print('Mean Squared Error: ', mean_squared_error(y_test, y_pred))
print('R2: ', r2_score(y_test, y_pred))
```
