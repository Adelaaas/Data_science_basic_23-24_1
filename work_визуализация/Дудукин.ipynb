````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('https://raw.githubusercontent.com/Adelaaas/Data_science_basic_23-24_1/main/work_%D0%B2%D0%B8%D0%B7%D1%83%D0%B0%D0%BB%D0%B8%D0%B7%D0%B0%D1%86%D0%B8%D1%8F/train_var_1.csv')
df.head()

plt.figure(figsize=(10, 6))
plt.scatter(df['Vmag'], df['e_Plx'], alpha=0.5)
plt.title('Диаграмма разброса между Vmag и e_Plx')
plt.xlabel('Vmag')
plt.ylabel('e_Plx')
plt.grid(True)
plt.show()
print("вывод: в основоном значения от 0 до 2.5 по e_plx и 0-11 по Vmag")

plt.figure(figsize=(10, 6))
cmap = sns.cubehelix_palette(as_cmap=True, light=.9)

sns.heatmap(df.isna().transpose(), cmap=cmap,
            cbar_kws={'label': 'Missing Data'})
plt.show()
print("conclusion: no missing data")

fig, axes = plt.subplots(nrows=1, ncols=7, figsize=(15, 5))

numeric_columns = df.select_dtypes(include='number')
for i, column in enumerate(numeric_columns.columns):

    axes[i].boxplot(df[column])
    axes[i].set_title(column)
    axes[i].set_ylabel(column)


plt.tight_layout()
plt.show()


print("у выбросов левее 5")
print("у plx много выбросов слева и справа")
print("у e-plx много выбросов справа > 2")
print("у B-V много выбросов справа")
print("у Amag много выбросов слева и справа")
print("у TargetClass вообще нет выбросов и все значения находятся в медиане")


plt.figure(figsize=(8, 5))
sns.barplot(x=df["TargetClass"], y=df['TargetClass'].value_counts())
plt.title('Распределение по TargetClass')
plt.xlabel('TargetClass')
plt.ylabel('Количество')
plt.show()
print("conclusion: 1 и 0 примерно одинаковое количество")


numeric_columns = df.select_dtypes(include=["number"])
plt.figure(figsize=(12,10), dpi=80)
sns.heatmap(df.select_dtypes(include=["number"]).corr(), xticklabels=df.select_dtypes(include=["number"]).corr().columns, yticklabels=df.select_dtypes(include=["number"]).corr().columns, cmap='RdYlGn', center=0, annot=True)
plt.title('Correlations', fontsize=22)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
print("между b-v и targetclass корреляция сильно отрицательно, остальные коррелируют мало")

plt.figure(figsize=(12,10))
sns.pairplot(df.select_dtypes(include=["number"]))
plt.show()
print("у всех похожие паттерны где есть основная часть в ограничение, а с targetclass есть зависимость по сайдам")


plt.figure(figsize=(8, 5))
sns.histplot(df['Vmag'], bins=30, kde=True, color='skyblue')
plt.title('Распределение переменной Vmag')
plt.xlabel('Vmag')
plt.ylabel('Частота')
plt.show()
print("основная частота между 6 и 10. максимум достигается гдето в 8")
````
