```python
import pandas as pd
import matplotlib.pyplot as plt
import random
import seaborn as sns
import numpy as np

a = pd.read_csv('https://raw.githubusercontent.com/Adelaaas/Data_science_basic_23-24_1/main/work_%D0%B2%D0%B8%D0%B7%D1%83%D0%B0%D0%BB%D0%B8%D0%B7%D0%B0%D1%86%D0%B8%D1%8F/train_var_2.csv')
x = a["age"]
y = a["balance"]
plt.scatter(x, y)
plt.xlabel("age")
plt.ylabel("balance")
plt.title("Точечный график по age и balance")
plt.show()
```

Можно сделать вывод, что значение balance не сильно зависит от значения age, оставаясь одинаковым примерно на всей возрастной оси, уменьшаясь только после значения 80.

```python
cmap = sns.cubehelix_palette(as_cmap=True, light=.9)

sns.heatmap(a.isna().transpose(), cmap=cmap,
            cbar_kws={'label': 'Пропущенные значения'})
```

Во всё массиве данных нет пропущенных значений

```python
plt.figure(figsize=(25, 10))
plt.subplot(1, 7, 1)
b = a["age"]
b.plot(kind='box')
plt.subplot(1, 7, 2)
b = a["balance"]
b.plot(kind='box')
plt.subplot(1, 7, 3)
b = a["day"]
b.plot(kind='box')
plt.subplot(1, 7, 4)
b = a["duration"]
b.plot(kind='box')
plt.subplot(1, 7, 5)
b = a["campaign"]
b.plot(kind='box')
plt.subplot(1, 7, 6)
b = a["pdays"]
b.plot(kind='box')
plt.subplot(1, 7, 7)
b = a["previous"]
b.plot(kind='box')
```

Медиана переменной age - чуть меньше 40, day - 15, previous - чуть больше нуля. У переменной day нет значений выбросов, но их множество у переменной pdays.

```python
yd = a["deposit"].value_counts()
xd = a["deposit"].unique()
sns.barplot(y = yd, x = xd)
```

Значение no в переменной deposite более распространенно, чем значение yes

```python
sns.heatmap(a.corr(), xticklabels=a.corr().columns, yticklabels=a.corr().columns, cmap='RdYlGn', center=0, annot=True)
```

Вывод: наибольшие корелляции в датафрейме - это положительная корелляция между переменными previous и pdays и отрицательная - между переменными pdays и campaign

```python
sns.pairplot(a, kind="reg")
```
