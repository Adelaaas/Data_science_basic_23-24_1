```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('https://raw.githubusercontent.com/Adelaaas/Data_science_basic_23-24_1/main/work_%D0%B2%D0%B8%D0%B7'
                 '%D1%83%D0%B0%D0%BB%D0%B8%D0%B7%D0%B0%D1%86%D0%B8%D1%8F/train_var_1.csv')
# print(df.head())
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Vmag'], df['e_Plx'])
plt.xlabel('Vmag')
plt.ylabel('e_Plx')
plt.title('Диаграмма разброса')
plt.show()
print("большинство значение лежат между 0 и 10 Vmag и между 1 и 2 e")
```

```python
plt.figure(figsize=(15,7))
plt.title('Тепловая карта пропущенных значений')
cmap = sns.cubehelix_palette(as_cmap=True, light=.9)
sns.heatmap(df.isna().transpose(), cmap=cmap,
            cbar_kws={'label': 'Missing Data'})
plt.show()
print("пропущенных значений нет")
```

```python
cols = df.select_dtypes(include='number').columns

plt.figure(figsize=(15, 20))
k = 1
for i in cols:
    if k > 4:
        plt.show()
        plt.figure(figsize=(15, 20))
        k = 1
    plt.subplot(2, 2, k)
    sns.boxplot(x=df[i])
    plt.title(f'Boxplot для {i}')
    k += 1
plt.show()


print("у vmag много выбросов левее 5")
print("у plx много выбросов слева и справа")
print("у e-plx много выбросов справа > 2")
print("у B-V много выбросов справа")
print("у Amag много выбросов слева и справа")
print("у TargetClass вообще нет выбросов и все значения находятся в медиане")
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='TargetClass', data=df)
plt.title("Распределение")
plt.xlabel("TargetClass")
plt.ylabel("Кол-во")
plt.show()
print("одинаковое")
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.select_dtypes(include=['number']).corr(),
            xticklabels=df.select_dtypes(include=['number']).corr().columns,
            yticklabels=df.select_dtypes(include=['number']).corr().columns, cmap='RdYlGn', center=0,
            annot=True)
plt.title('Correlogram of all values', fontsize=22)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()

print("везде, кроме targetClass и B-V корреляция почти отсуствует.")
print("у targetClass и B-V корреляция сильно отрицательная")
```

```python
plt.figure(figsize=(12, 10))
sns.pairplot(df.select_dtypes(include=["number"]))
plt.show()

print("сложно, времени нет")
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['B-V'], bins=30, kde=True, color='skyblue')
plt.title('Распределение переменной B-V')
plt.xlabel('B-V')
plt.ylabel('Частота')
plt.show()

print("выбросов почти нет. разброс небольшой.")
```
