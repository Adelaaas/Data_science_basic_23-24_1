````python

import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('https://raw.githubusercontent.com/Adelaaas/Data_science_basic_23-24_1/main/work_%D0%B2%D0%B8%D0%B7%D1%83%D0%B0%D0%BB%D0%B8%D0%B7%D0%B0%D1%86%D0%B8%D1%8F/train_var_3.csv')
df.head()
fig, ax = plt.subplots()
ax.scatter(df["MinTemp"], df["MaxTemp"])
ax.set_ylabel("MaxTemp")
ax.set_xlabel("MinTemp")
plt.title("Диаграмма разброса")
plt.show()
````

````python
Вывод: Между MaxTemp и MinTemp есть линейная зависимость.
````

````python
import seaborn as sns
plt.figure(figsize = (15,10))
cmap = sns.cubehelix_palette(as_cmap = True, light = .9)
sns.heatmap(df.isna().transpose(), cmap=cmap, cbar_kws = {"label": "Missing Data"})
plt.title("График анализа пропущенных значений")
plt.show()
````

````python
Вывод: Больше всего пропущенных значений имеют sunshine и cloud3pm.
````

````python
fig, axes = plt.subplots(1, 16, figsize=(40,20))
sns.boxplot(df["MinTemp"], ax=axes[0], width=1)
sns.boxplot(df["MaxTemp"], ax=axes[1], width=1)
sns.boxplot(df["Rainfall"], ax=axes[2], width=1)
sns.boxplot(df["Evaporation"], ax=axes[3], width=1)
sns.boxplot(df["Sunshine"], ax=axes[4], width=1)
sns.boxplot(df["WindGustSpeed"], ax=axes[5], width=1)
sns.boxplot(df["Humidity9am"], ax=axes[6], width=1)
sns.boxplot(df["Humidity3pm"], ax=axes[7], width=1)
sns.boxplot(df["Cloud9am"], ax=axes[8], width=1)
sns.boxplot(df["Cloud3pm"], ax=axes[9], width=1)
sns.boxplot(df["Pressure9am"], ax=axes[10], width=1)
sns.boxplot(df["Pressure3pm"], ax=axes[11], width=1)
sns.boxplot(df["Temp9am"], ax=axes[12], width=1)
sns.boxplot(df["Temp3pm"], ax=axes[13], width=1)
sns.boxplot(df["RainToday"], ax=axes[14], width=1)
sns.boxplot(df["RainTomorrow"], ax=axes[15], width=1)

axes[0].set_xlabel("MinTemp")
axes[1].set_xlabel("MaxTemp")
axes[2].set_xlabel("Rainfall")
axes[3].set_xlabel("Evaporation")
axes[4].set_xlabel("Sunshine")
axes[5].set_xlabel("WindGustSpeed")
axes[6].set_xlabel("Humidity9am")
axes[7].set_xlabel("Humidity3pm")
axes[8].set_xlabel("Cloud9am")
axes[9].set_xlabel("Cloud3pm")
axes[10].set_xlabel("Pressure9am")
axes[11].set_xlabel("Pressure3pm")
axes[12].set_xlabel("Temp9am")
axes[13].set_xlabel("Temp3pm")
axes[14].set_xlabel("RainToday")
axes[15].set_xlabel("RainTomorrow")
plt.title('boxplot')
plt.show()
````

````python
Вывод:Минимальное значение: Sunshine, WindGustSpeed, Humidity3pm, Cloud9am,C loud3pm; Максимальное значение: Sunshine, Humidity9am, Humidity3pm,Cloud9am, Cloud3pm.
````

````python
plt.figure(figsize = (10, 5))
sns.barplot(x = df["RainTomorrow"], y = df["RainTomorrow"].value_counts())
plt.title("barplot по переменной RainTomorrow")
plt.ylabel("Количество")
plt.xlabel("RainTomorrow")
plt.show()
````

````python
Вывод: 0 больше.
````

````python
numeric_columns = df.select_dtypes(include=["number"])
plt.figure(figsize = (15, 10))
sns.heatmap(df.select_dtypes(include=["number"]).corr(), xticklabels=df.select_dtypes(include=["number"]).corr().columns, yticklabels=df.select_dtypes(include=["number"]).corr().columns, cmap='RdYlGn', center=0, annot=True)
plt.title("Матрица корреляции для всех переменных")
plt.show()
````

````python
plt.figure(figsize= (10, 5))
plt.hist(df["Sunshine"])
plt.title("График распределения переменной Sunshine")
plt.xlabel("Sunshine")
plt.ylabel("Частота")
plt.show()
````

````python
Вывод: Минимальная частота: 12-14; Максимальная частота: 10-12.
````
