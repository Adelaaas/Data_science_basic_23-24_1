```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv('https://raw.githubusercontent.com/Adelaaas/Data_science_basic_23-24_1/main/work_%D0%B2%D0%B8%D0%B7%D1%83%D0%B0%D0%BB%D0%B8%D0%B7%D0%B0%D1%86%D0%B8%D1%8F/train_var_2.csv')

fig, ax = plt.subplots()

ax.scatter(df['age'], df['balance'], c = 'deeppink')
plt.title("точечная зависимочть age и balance")
plt.xlabel("age")
plt.ylabel("balance")

plt.show()
```

balance примерно 0, есть выбросы вверх

```python
fig, axes = plt.subplots(1,7, figsize=(25,7))
axes[0].set_title("boxplot для age")
sns.boxplot(df['age'], ax=axes[0], width=0.3)
plt.grid(True)

axes[1].set_title("boxplot для balance")
sns.boxplot(df['balance'], ax=axes[1], width=0.3)

axes[2].set_title("boxplot для day")
sns.boxplot(df['day'], ax=axes[2], width=0.3)

axes[3].set_title("boxplot для duration")
sns.boxplot(df['duration'], ax=axes[3], width=0.3)

axes[4].set_title("boxplot для campaign")
sns.boxplot(df['campaign'], ax=axes[4], width=0.3)

axes[5].set_title("boxplot для pdays")
sns.boxplot(df['pdays'], ax=axes[5], width=0.3)

axes[6].set_title("boxplot для previous")
sns.boxplot(df['previous'], ax=axes[6], width=0.3)
```

```python
df.isna().sum()
plt.figure(figsize=(15,7))
plt.title("график анализа пропущенных значений")
cmap = sns.cubehelix_palette(as_cmap=True, light=.9)

sns.heatmap(df.isna().transpose(), cmap=cmap,
            cbar_kws={'label': 'Missing Data'})
```
